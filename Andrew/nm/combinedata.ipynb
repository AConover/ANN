```python
from os import listdir
from os.path import isfile, join
import pandas as pd
```

```python
path = "Results"+"/"
onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]
df = pd.DataFrame()
for file in onlyfiles:
    cff = pd.read_csv(path+file)
    df = df.append(cff)
```

```python
means = pd.DataFrame()
for x in df['set'].unique():
    arr = df[df['set']==x]
    for i in arr['replica'].unique():
        s = arr[arr['replica']==i]
        s = s.mean()
        means = means.append(s, ignore_index=True)
means = means[['set','replica','ReH','ReE','ReHT']]
```

```python
means.to_csv('combinedresults.csv')
```
