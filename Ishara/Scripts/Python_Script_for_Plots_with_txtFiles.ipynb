```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#t1=pd.read_csv('./Results_txts/Dustin_Result.txt',sep='\s+')
txt_files_path='./Results_txts/'
### Here you can define the files using for combined plots

tfn1=txt_files_path+'Dustin_Result.txt'
tfn2=txt_files_path+'Test_result_Opt_Adam_Act_selu_Rep_1000_LR_0-02.txt'
tfn3=txt_files_path+'Test_result_Opt_Adam_Act_selu_Rep_1000_LR_0-5.txt'
tfn4=txt_files_path+'Test_result_Opt_Adam_Act_tanh_Rep_1000_LR_0-02.txt'
tfn5=txt_files_path+'Test_result_Opt_Adam_Act_tanh_Rep_1000_LR_0-5.txt'
tfn6=txt_files_path+'Test_result_Opt_Adamax_Act_selu_Rep_1000_LR_0-02.txt'
tfn7=txt_files_path+'Test_result_Opt_Adamax_Act_selu_Rep_1000_LR_0-5.txt'

txt_files=(tfn1,tfn2,tfn3,tfn4,tfn5,tfn6,tfn7)
```

```python
def FileData(filename,CFF,dCFF,clr,lbl,shift):
    tempdf=pd.read_csv(filename,sep='\s+')
    tempSet=np.array(tempdf["Set"])+shift
    tempCFF=tempdf[CFF]
    temp_dCFF=tempdf[dCFF]
    plt.errorbar(tempSet,tempCFF,temp_dCFF,fmt='o',color=clr,label=lbl)

#FileData(txt_files[1],"ReE","dReE",'r','',0.2)

f1=plt.figure(1,figsize=(20,10))
plt.title("ReE (1000 Replicas)",fontsize=25)
FileData(txt_files[0],"ReE_data","dReE_data",'red','Data(Actual)',0.0)
FileData(txt_files[0],"ReE","dReE",'orange','Dustin',0.1)
FileData(txt_files[1],"ReE","dReE",'blue','Opt_Adam_Act_selu_LR_0-02',0.2)
FileData(txt_files[2],"ReE","dReE",'green','Opt_Adam_Act_selu_LR_0-5',0.3)
FileData(txt_files[3],"ReE","dReE",'gold','Opt_Adam_Act_tanh_LR_0-02',0.4)
FileData(txt_files[4],"ReE","dReE",'magenta','Opt_Adam_Act_tanh_LR_0-5',0.5)
FileData(txt_files[3],"ReE","dReE",'deepskyblue','Opt_Adamax_Act_selu_LR_0-02',0.6)
FileData(txt_files[4],"ReE","dReE",'chocolate','Opt_Adamax_Act_selu_LR_0-5',0.7)
plt.ylim(-120,-10)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.legend(loc=4,fontsize=20,handlelength=3)



f2=plt.figure(2,figsize=(20,10))
plt.title("ReH (1000 Replicas)",fontsize=25)
FileData(txt_files[0],"ReH_data","dReH_data",'red','Data(Actual)',0.0)
FileData(txt_files[0],"ReH","dReH",'orange','Dustin',0.1)
FileData(txt_files[1],"ReH","dReH",'blue','Opt_Adam_Act_selu_LR_0-02',0.2)
FileData(txt_files[2],"ReH","dReH",'green','Opt_Adam_Act_selu_LR_0-5',0.3)
FileData(txt_files[3],"ReH","dReH",'gold','Opt_Adam_Act_tanh_LR_0-02',0.4)
FileData(txt_files[4],"ReH","dReH",'magenta','Opt_Adam_Act_tanh_LR_0-5',0.5)
FileData(txt_files[3],"ReH","dReH",'deepskyblue','Opt_Adamax_Act_selu_LR_0-02',0.6)
FileData(txt_files[4],"ReH","dReH",'chocolate','Opt_Adamax_Act_selu_LR_0-5',0.7)
plt.ylim(0,25)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.legend(loc=1,fontsize=20,handlelength=3)



f3=plt.figure(3,figsize=(20,10))
plt.title("ReHtilde (1000 Replicas)",fontsize=25)
FileData(txt_files[0],"ReHtilde_data","dReHtilde_data",'red','Data(Actual)',0.0)
FileData(txt_files[0],"ReHtilde","dReHtilde",'orange','Dustin',0.1)
FileData(txt_files[1],"ReHtilde","dReHtilde",'blue','Opt_Adam_Act_selu_LR_0-02',0.2)
FileData(txt_files[2],"ReHtilde","dReHtilde",'green','Opt_Adam_Act_selu_LR_0-5',0.3)
FileData(txt_files[3],"ReHtilde","dReHtilde",'gold','Opt_Adam_Act_tanh_LR_0-02',0.4)
FileData(txt_files[4],"ReHtilde","dReHtilde",'magenta','Opt_Adam_Act_tanh_LR_0-5',0.5)
FileData(txt_files[3],"ReHtilde","dReHtilde",'deepskyblue','Opt_Adamax_Act_selu_LR_0-02',0.6)
FileData(txt_files[4],"ReHtilde","dReHtilde",'chocolate','Opt_Adamax_Act_selu_LR_0-5',0.7)
plt.ylim(2,12)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.legend(loc=1,fontsize=20,handlelength=3)



f4=plt.figure(4,figsize=(20,10))
plt.title("F values (1000 Replicas)",fontsize=25)
FileData(txt_files[1],"F2_data","dF2_data",'red','Data(Actual)',0.0)
FileData(txt_files[1],"F2","dF2",'blue','Opt_Adam_Act_selu_LR_0-02',0.2)
FileData(txt_files[2],"F2","dF2",'green','Opt_Adam_Act_selu_LR_0-5',0.3)
FileData(txt_files[3],"F2","dF2",'gold','Opt_Adam_Act_tanh_LR_0-02',0.4)
FileData(txt_files[4],"F2","dF2",'magenta','Opt_Adam_Act_tanh_LR_0-5',0.5)
FileData(txt_files[3],"F2","dF2",'deepskyblue','Opt_Adamax_Act_selu_LR_0-02',0.6)
FileData(txt_files[4],"F2","dF2",'chocolate','Opt_Adamax_Act_selu_LR_0-5',0.7)
plt.ylim(0,0.16)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.legend(loc=1,fontsize=20,handlelength=3)



import matplotlib.backends.backend_pdf
pdf = matplotlib.backends.backend_pdf.PdfPages("SummaryPlot.pdf")
for fig in range(1,5):
    pdf.savefig(fig)
pdf.close()
```
